```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df.drop('Car_Name',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Age']=2024-df['Year']
```

```python
df.drop('Year',axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop('Selling_Price',axis=1)
y=df['Selling_Price']
```

```python
X.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder



# Create the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
         # Standardize numerical features
        ('ohe', OneHotEncoder(sparse_output=False), [2,3,4])   # One-hot encode categorical features
    ],
    remainder=StandardScaler()  
).set_output(transform='pandas')

```

```python
X=preprocessor.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=52)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
from sklearn import metrics
```

```python
y_pred=lr.predict(X_test)
```

```python
metrics.r2_score(y_test,y_test)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
r=Ridge(alpha=10)
```

```python
r.fit(X_train,y_train)
```

```python
y_pred=r.predict(X_test)
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
l=Lasso(alpha=0.1)
```

```python
l.fit(X_train,y_train)
```

```python
y_pred=l.predict(X_test)
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=200,n_jobs=5)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred=rfr.predict(X_test)
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

## Gradient Boosting regression

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(n_estimators=150,learning_rate=0.1)
```

```python
gbr.fit(X_train,y_train)
```

```python
y_pred=gbr.predict(X_test)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
def predict():
    input_values=[]
   #current_price_entry
    current_price=float(app.current_price_entry.get())
    input_values.append(current_price)

#Kms_Driven_entry
    
    Kms_Drivenx=float(app.Kms_Driven_entry.get())
    input_values.append(Kms_Drivenx)

#Fuel_Type_var
    Fuel_Type=app.Fuel_Type_var.get()
    input_values.append(Fuel_Type)
#Seller_Type_var
    Seller_Type=app.Seller_Type_var.get()
    input_values.append(Seller_Type)
#Transmission_var
    Transmission=app.Transmission_var.get()
    input_values.append(Transmission)
#  Owner_entry  
    Owner=float(app.Owner_entry.get())
    input_values.append(Owner)
#   Age_entry
    Age=float(app.Age_entry.get())
    input_values.append(Age)
  
    input_values = [input_values]
    app.output_entry.delete(0,tk.END)
    print(input_values)
    observation = preprocessor.transform(input_values)
    prediction = gbr.predict(observation)[0]
    app.output_entry.insert(tk.END,prediction)

```

```python
import tkinter as tk
from tkinter import ttk
class Model:
    def __init__(self, root):
        self.root = root
        self.root.title("Model Deployment with tkinter")

        # Create a main frame to hold all the widgets
        self.main_frame = tk.Frame(self.root)
        self.main_frame.pack(padx=10, pady=10)

        # Create a dictionary to store the user input
        self.user_input = {}

        # Current price
        tk.Label(self.main_frame, text="Enter Current Price (in lacs):").grid(row=0, column=0, padx=5, pady=(5, 10))
        self.current_price_entry = tk.Entry(self.main_frame, width=20)
        self.current_price_entry.grid(row=0, column=1, padx=5, pady=5)

        

        # Kms_Driven
        tk.Label(self.main_frame, text="kms Driven:").grid(row=1, column=0, padx=5, pady=(5, 10))
        self.Kms_Driven_entry = tk.Entry(self.main_frame, width=20)
        self.Kms_Driven_entry.grid(row=1, column=1, padx=5, pady=5)

        # Fuel_Type
        tk.Label(self.main_frame, text="Choose Fuel Type:").grid(row=2, column=0, padx=5, pady=(5, 10))
        self.Fuel_Type_var = tk.StringVar()
        Fuel_Type_combobox = ttk.Combobox(self.main_frame, textvariable=self.Fuel_Type_var)
        Fuel_Type_combobox['values'] = ('Petrol', 'Diesel', 'CNG')
        Fuel_Type_combobox.current(0)
        Fuel_Type_combobox.grid(row=2, column=1, padx=5, pady=5)

        # Seller_Type
        tk.Label(self.main_frame, text="Choose Seller Type:").grid(row=3, column=0, padx=5, pady=(5, 10))
        self.Seller_Type_var = tk.StringVar()
        Seller_Type_combobox = ttk.Combobox(self.main_frame, textvariable=self.Seller_Type_var)
        Seller_Type_combobox['values'] = ('Dealer', 'Individual')
        Seller_Type_combobox.current(0)
        Seller_Type_combobox.grid(row=3, column=1, padx=5, pady=5)

        # Transmission
        tk.Label(self.main_frame, text="Choose Transmission:").grid(row=4, column=0, padx=5, pady=(5, 10))
        self.Transmission_var = tk.StringVar()
        Transmission_combobox = ttk.Combobox(self.main_frame, textvariable=self.Transmission_var)
        Transmission_combobox['values'] = ('Manual', 'Automatic')
        Transmission_combobox.current(0)
        Transmission_combobox.grid(row=4, column=1, padx=5, pady=5)

        # Owner
        tk.Label(self.main_frame, text="Enter Owners numbers:").grid(row=5, column=0, padx=5, pady=(5, 10))
        self.Owner_entry = tk.Entry(self.main_frame, width=20)
        self.Owner_entry.grid(row=5, column=1, padx=5, pady=5)

        # Age
        tk.Label(self.main_frame, text="Enter Age:").grid(row=6, column=0, padx=5, pady=(5, 10))
        self.Age_entry = tk.Entry(self.main_frame, width=20)
        self.Age_entry.grid(row=6, column=1, padx=5, pady=5)

        # Prediction Output
        label = tk.Label(self.root, text="Prediction Output:")
        label.pack()
        self.output_entry = tk.Entry(self.root, width=40)
        self.output_entry.pack()

        self.button = tk.Button(self.root, text="Predict", command=predict)
        self.button.pack()

root = tk.Tk()
app = Model(root) 
root.mainloop()

```
